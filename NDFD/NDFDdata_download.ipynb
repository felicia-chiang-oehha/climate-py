### Download NDFD data

This script downloads requested data from NCEI.

```python
# import libraries
import requests
import numpy as np
import pandas as pd
```

```python
# save to folder
wd = r"K:\\Data\\NDFD\\"

# address to download from
web_address = 'https://www.ncei.noaa.gov/pub/has/HAS012581603/'
```

```python
# create list of dates
datelist = pd.date_range(start = '20150101', end = '20241231').strftime('%Y%m%d').tolist()
datelist
```

```python
#create url list
currentfile_subname = '9959_NDFD_YTQ_'

urllist = [web_address + currentfile_subname + date + '.tar' for date in datelist]

#create file name list
filenamelist = [wd + currentfile_subname + date + '.tar' for date in datelist]

#download...
for filenum in np.arange(0, len(filenamelist)):
    #get the url
    response = requests.get(urllist[filenum])
    with open(filenamelist[filenum], 'wb') as file:
        file.write(response.content)
    print(f'{filenamelist[filenum]} downloaded.')
```
